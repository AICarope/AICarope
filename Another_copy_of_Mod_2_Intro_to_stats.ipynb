# Descriptive statistics

Method to summarize and describe the main features of a dataset


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = [10, 12, 23, 23, 16, 23, 21, 16, 19, 12]

# Example data
data_2d = pd.DataFrame({
    "Variable 1": [10, 12, 23, 23, 16, 23, 21, 16, 19, 12],
    "Variable 2": [15, 18, 25, 24, 22, 21, 19, 17, 18, 20]
})

data_2d
```

### Measures of central tendency

```python
# Mean
# Arithmetic average of the data.
# Calculated by adding all the values and dividing them by the number of values

mean_value = np.mean(data)
mean_value
```

```python
# Median
# Middle value when the data points are arranged in order

median_value = np.median(data)
median_value
```

```python
# Mode
# The most frequently occurring value in the dataset

mode_value = pd.Series(data).mode()[0]  # Mode can return multiple values, we take the first
mode_value
```

### Measutes of variability (dispersion)

```python
# Range
# The difference between the maximum and minimum values

range_value = np.ptp(data)  # ptp() returns the range (max - min)
range_value
```

```python
# Variance
# A measure of how much the data points differ from the mean on average

variance_value = np.var(data, ddof=0)  # Population variance
variance_value
```

```python
# Standard deviation
# Square root of the variance, giving a sense of how much data points typically deviate from the mean in the same units as the data

std_dev_value = np.std(data, ddof=0)  # Population standard deviation
std_dev_value
```

```python
np.sqrt(22.65)
```

```python
# Interquartile Range (IQR)
q75, q25 = np.percentile(data, [75, 25])
iqr_value = q75 - q25
iqr_value
```

```python
# Collecting results in a dataframe
summary_df = pd.DataFrame({
    "Statistic": ["Mean", "Median", "Mode", "Range", "Variance", "Standard Deviation", "Interquartile Range"],
    "Value": [mean_value, median_value, mode_value, range_value, variance_value, std_dev_value, iqr_value]
})

summary_df
```

### Graphical representations

```python
# Histogram
data_2d["Variable 1"].plot(kind='hist', bins=5, edgecolor='black', title='Histogram of Variable 1')
plt.xlabel('Value Ranges')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram
data_2d["Variable 2"].plot(kind='hist', bins=5, edgecolor='black', title='Histogram of Variable 2')
plt.xlabel('Value Ranges')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box Plot
data_2d.plot(kind='box', vert=False, title='Box Plot of Variable 1 and Variable 2')
plt.xlabel('Values')
plt.show()
```

```python
# Scatter Plot between Variable 1 and Variable 2
data_2d.plot(kind='scatter', x='Variable 1', y='Variable 2', title='Scatter Plot of Variable 1 vs Variable 2')
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.show()
```

# Example: Normal Distribution in Python

```python
# Parameters for the normal distribution
mean = 0
std_dev = 1
```

```python
# Generating random data from a normal distribution
data = np.random.normal(mean, std_dev, 1000)
```

```python
data
```

```python
# Visualizing the normal distribution using seaborn and matplotlib
plt.figure(figsize=(8, 6))
sns.histplot(data, kde=True, color='blue', bins=30)
plt.title('Normal Distribution with Mean = 0, Std Dev = 1')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
